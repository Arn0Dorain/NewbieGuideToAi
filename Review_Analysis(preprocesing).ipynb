```
f = open('/content/drive/MyDrive/Colabfiles/reviews.txt','r')#opening files from Gdrive into our code 
reviews = list(map(lambda x:x[:-1],f.readlines()))
f.close()

f = open('/content/drive/MyDrive/Colabfiles/labels.txt','r')#has to be labbled to reviews
labels = list(map(lambda x:x[:-1].upper(),f.readlines()))
f.close()

def printlabl_rev(i):#function to print lables corresponds to review
    print(labels[i] + "\t:\t" + reviews[i][:80] + "...") 
```

```
from collections import Counter #Holds the data from sets 
import numpy as np
```

```
count_positive = Counter()
count_negative = Counter()
total_count = Counter()
```

```
#using for to sort postive and negative
for i in range(len(reviews)):
  if (labels[i]=='POSITIVE') :
    for j in reviews[i].split(" "):
      count_positive[j] +=1
      total_count[j] +=1
  else:
    for j in reviews[i].split(" "):
      count_negative[j] +=1
      total_count[j] +=1
```

```
pos_neg_rats = Counter()
for t,cnt in list(total_count.most_common()):
  if (cnt>154) :
    pn_rat = count_positive[t] / float(count_negative[t]+1)
    pos_neg_rats[t] = pn_rat

for word,ratio in pos_neg_rats.most_common():
  if(ratio>1):
    pos_neg_rats[word] = np.log(ratio)
  else:
    pos_neg_rats[word] = -np.log((1 / (ratio+0.01)))
```

Preparing input and output data


```
vocab = set(total_count.keys())
print(len(vocab))
```

```
import numpy as np

layer_0 = np.zeros((1,len(vocab)))
layer_0
```

```
word2index ={}
for i,word in enumerate(vocab):
  word2index[word] =i
word2index
```

```
def update_iplayer(review):
   global layer_0
   layer_0 += 0
   for word in review.split(" "):
     layer_0[0][word2index[word]] += 1

update_iplayer(reviews[0])
```
